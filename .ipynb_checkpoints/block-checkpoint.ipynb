清单7.1 通过 SHA-256 对“hello”进行哈希处理

```python
import hashlib

hash_hello = hashlib.sha256(b"hello").hexdigest()

print(hash_hello)
```

清单7.2 通过 SHA-256 对“hallo”进行哈希处理

```python
import hashlib

hash_hello = hashlib.sha256(b"hello").hexdigest()
hash_hallo = hashlib.sha256(b"hallo").hexdigest()
print(hash_hello)
print(hash_hallo)
```

清单7.3 通过 SHA-256 对各种字符串进行哈希处理

```python
import hashlib

hash_hello = hashlib.sha256(b"hello").hexdigest()
hash_hallo = hashlib.sha256(b"hallo").hexdigest()
hash_helloworld = hashlib.sha256(b"hello world!").hexdigest()

print(hash_hello)
print(hash_hallo)
print(hash_helloworld)
```
